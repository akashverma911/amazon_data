```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import re
import time
import json
```

```python
df = pd.read_excel("amazon_data.xlsx")
```

```python
HEADERS = ({'User-Agent':
           'Mozilla/5.6 (X11; Linux x86_64)AppleWebKit/537.36 (KHTML, like Gecko)Chrome/44.0.2403.157 Safari/537.36',
                           'Accept-Language': 'en-US, en;q=0.5'})
data=[]
timespent=[]



for i in range(0,1000,100):
    start_time=time.time()    
    for j in range(i,i+100,1):
        url = "https://www.amazon." + str(df["country"][j]) + "/dp/" + str(df["Asin"][j])
        r = requests.get(url,headers=HEADERS)
        soup = BeautifulSoup(r.content, 'lxml')
        print("---------------------------------------------------")
        print(url)
        
        data_row={}
    # -------------------------------------------------------------------------------------------------------------
        try:
            title = soup.find("span",attrs={"id": 'productTitle'}) 
            data_row['title']=title.text      
        except AttributeError:
            data_row['title']= None
        
    # ---------------------------------------------------------------------------------------------------------------    
        try:
            image = soup.find("div",attrs={"id": 'imgTagWrapperId'})  
            data_row['img']=image.img['src']      
        except AttributeError:
            try:
                image = soup.find("div",attrs={"id": 'img-canvas'})
                data_row['img']=image.img['src']   
            except AttributeError:
                data_row['img']= "NA"
        
    # ------------------------------------------------------------------------------------------------------------------    
        try:
            price = soup.find("span",attrs={"class": 'a-offscreen'})  
            data_row['price']=price.text  
        except AttributeError:
            data_row['price']= "NA"
        
    # ---------------------------------------------------------------------------------------------------------------------
        try:
            detail = soup.find("ul",attrs={"class": 'detail-bullet-list'})  
            s = detail.text
            s=re.sub(' +', ' ', s)
            s = s.replace('\n',"")
            s = s.replace('\u200e',"")
            data_row['detail']=s
        except AttributeError:
            try:
                detail = soup.find("div",attrs={"id": 'featurebullets_feature_div'})  
                s = detail.text
                s=re.sub(' +', ' ', s)
                s = s.replace('\n',"")
                s = s.replace('\u200e',"")
                data_row['detail']=s
            except AttributeError:
                data_row['detail']= "NA"
        
        print(data_row)
        data.append(data_row)
    timespent.append(time.time()-start_time)
dff = pd.DataFrame(data)
print(dff)
```

```python
dff.dropna(subset = ["title"], inplace=True)
print(dff)
```

```python
dff.to_csv (r'export_dataframe.csv', index = False, header=True)
json_data = dff.to_json(orient = 'columns')
print(json_data)
```
